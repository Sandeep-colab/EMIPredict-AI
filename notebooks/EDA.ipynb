```python
df['age'] = pd.to_numeric(df['age'], errors='coerce').fillna(0).astype(int)
df['gender'] = df['gender'].astype(str).str.lower().replace({'m': 'Male', 'male': 'Male', 'f': 'Female', 'female': 'Female'})
df.dropna(subset=['education'], inplace=True)
df['monthly_salary'] = pd.to_numeric(df['monthly_salary'], errors='coerce').fillna(int(round(pd.to_numeric(df['monthly_salary'], errors='coerce').median()))).astype('Int64')
df['monthly_rent'] = df['monthly_rent'].replace(0, np.nan).fillna(df['monthly_rent'].replace(0, np.nan).median()).astype(int)
df['school_fees'] = df['school_fees'].replace(0, np.nan).fillna(df['school_fees'].replace(0, np.nan).median()).astype(int)
df['college_fees'] = df['college_fees'].replace(0, np.nan).fillna(df['college_fees'].replace(0, np.nan).median()).astype(int)
df['travel_expenses']=df['travel_expenses'].astype(int)
df['groceries_utilities']=df['groceries_utilities'].astype(int)
df['other_monthly_expenses']=df['other_monthly_expenses'].astype(int)
df['current_emi_amount'] = df['current_emi_amount'].replace(0, np.nan).fillna(df['current_emi_amount'].replace(0, np.nan).median()).astype(int)
df['credit_score'] = df['credit_score'].replace(0, np.nan)
df.loc[(df['credit_score'] < 300) | (df['credit_score'] > 850), 'credit_score'] = np.nan
df['credit_score'] = df['credit_score'].fillna(int(round(df['credit_score'].median()))).astype(int)
df['bank_balance'] = pd.to_numeric(df['bank_balance'], errors='coerce').fillna(pd.to_numeric(df['bank_balance'], errors='coerce').median()).astype(int)
df['emergency_fund'] = pd.to_numeric(df['emergency_fund'], errors='coerce').fillna(df['emergency_fund'].median()).astype(int)
df['requested_amount']=df['requested_amount'].astype(int)
df['requested_tenure']=df['requested_tenure'].astype(int)
df['max_monthly_emi']=df['max_monthly_emi'].astype(int)
```

```python
gender_mapping = {'Male': 1, 'Female': 0}
df['gender'] = df['gender'].map(gender_mapping)
```

```python
employment_type= {
    'Private': 1,
    'Government': 2,
    'Self-employed': 3
}

df['employment_type'] = df['employment_type'].map(employment_type)
```

```python
company_type= {
    'Mid-size': 1,
    'MNC': 2,
    'Startup': 3,
    'Large Indian':4,
    'Small':5
}

df['company_type'] = df['company_type'].map(company_type)
```

```python
house_mapping = {
    'Rented': 1,
    'Family': 2,
    'Own': 3
}

df['house_type'] = df['house_type'].map(house_mapping)
```

```python
existing_loans = {'Yes': 1, 'No': 0}
df['existing_loans'] = df['existing_loans'].map(existing_loans)
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['emi_eligibility'] = le.fit_transform(df['emi_eligibility'])
```

```python
# Use get_dummies for One-Hot Encoding
df = pd.get_dummies(df, columns=['emi_scenario'], prefix='emi_scenario', drop_first=True)
```

```python
# Create a total expenses column first
expense_cols = ['monthly_rent', 'school_fees', 'college_fees', 'travel_expenses', 
                'groceries_utilities', 'other_monthly_expenses', 'current_emi_amount']
df['total_monthly_expenses'] = df[expense_cols].sum(axis=1)
```

```python
# List of columns to drop (the six non-EMI individual expenses)
columns_to_drop = [
    'monthly_rent', 
    'school_fees', 
    'college_fees', 
    'travel_expenses', 
    'groceries_utilities', 
    'other_monthly_expenses'
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
# Calculate new financial ratios
df['DTI_ratio'] = df['current_emi_amount'] / df['monthly_salary']
df['ETI_ratio'] = df['total_monthly_expenses'] / df['monthly_salary']
df['free_cash_flow'] = df['monthly_salary'] - df['total_monthly_expenses']
```

```python
# Drop the original components that are now better represented by the ratios
columns_to_drop = [
    'current_emi_amount',
    'total_monthly_expenses'
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
# Handle division by zero/NaN (if monthly_salary was zero, replace inf/-inf with a large number or 0)
df['DTI_ratio'].replace([np.inf, -np.inf], 0, inplace=True)
df['ETI_ratio'].replace([np.inf, -np.inf], 0, inplace=True)
```

```python
import numpy as np # Ensure numpy is imported

# Calculate family-based features (add a small epsilon or 1 to avoid division by zero)
df['income_per_capita'] = df['monthly_salary'] / df['family_size'].replace(0, 1)
df['income_per_dependent'] = df['monthly_salary'] / (df['dependents'] + 1)
```

```python
# Drop the original columns that are now captured in the ratios
columns_to_drop = [
    'family_size',
    'dependents'
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
# Calculate job stability
df['job_stability_ratio'] = df['years_of_employment'] / df['age'].replace(0, 1)
```

```python
# Drop the original columns whose information is now captured in the ratio
columns_to_drop = [
    'years_of_employment',
    'age'
]

df.drop(columns=columns_to_drop, axis=1, inplace=True)
```

```python
# Identify the target variables
TARGET_CLASSIFICATION = 'emi_eligibility'  # The encoded categorical column
TARGET_REGRESSION = 'max_monthly_emi'      # The continuous numeric column
```

```python
# Create the Feature Matrix (X): Drop both target columns from the DataFrame
X = df.drop(columns=[TARGET_CLASSIFICATION, TARGET_REGRESSION], axis=1)
```

```python
# Create the two Target Vectors (y)
y_class = df[TARGET_CLASSIFICATION]
y_reg = df[TARGET_REGRESSION]
```

```python
print(f"Features (X) shape: {X.shape}")
print(f"Classification Target (y_class) shape: {y_class.shape}")
print(f"Regression Target (y_reg) shape: {y_reg.shape}")
```

```python
from sklearn.model_selection import train_test_split

# Split X and the classification target first (using stratify to handle imbalance)
X_train, X_test, y_class_train, y_class_test = train_test_split(
    X, y_class, test_size=0.2, random_state=42, stratify=y_class
)
```

```python
# Use the same indices to split the regression target
y_reg_train = y_reg.loc[y_class_train.index]
y_reg_test = y_reg.loc[y_class_test.index]
print("Data successfully split for both models.")
```

```python
from imblearn.combine import SMOTEENN

# Convert all numeric columns in the training features to standard float64.
# This ensures consistency and prevents the 'safe cast' error during resampling.
X_train_float = X_train.astype('float64')

# Apply SMOTEENN to the float-converted data
smote_enn = SMOTEENN(random_state=42)
X_class_resampled, y_class_resampled = smote_enn.fit_resample(X_train_float, y_class_train)


print("Data successfully converted to float64 and resampled.")
print(f"Classification training data resampled. New training size: {X_class_resampled.shape}")
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import classification_report
import xgboost as xgb

# ----------------- Model Initialization -----------------
models = {
    "XGBoost": xgb.XGBClassifier(
        objective='multi:softmax',
        num_class=3,
        random_state=42,
        use_label_encoder=False,
        eval_metric='mlogloss',
        n_estimators=100  # Default or small number for quick comparison
    ),
    "Random Forest": RandomForestClassifier(
        n_estimators=100,
        random_state=42,
        class_weight='balanced' # Helps with imbalance, though we already used SMOTE
    ),
    "Logistic Regression": LogisticRegression(
        multi_class='multinomial', # Required for multi-class problems
        solver='lbfgs',
        max_iter=500, # Increase iterations for convergence
        random_state=42
    ),
    "K-Nearest Neighbors": KNeighborsClassifier(
        n_neighbors=5 # Default number of neighbors
    )
}

X_test_float = X_test.astype('float64') # Ensure test data matches training type

# ----------------- Training and Evaluation Loop -----------------
for name, model in models.items():
    print(f"\n--- Training {name} ---")

    # Train using the SMOTEENN-resampled data
    model.fit(X_class_resampled, y_class_resampled)
    
    # Predict on the original test set
    y_pred = model.predict(X_test_float)
    
    # Evaluate performance
    print(f"\n{name} Performance on Test Set:")
    # Classification Report is crucial for imbalanced multi-class evaluation
    print(classification_report(y_class_test, y_pred, zero_division=0))
    
    print("-" * 50)
```

```python
import xgboost as xgb
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Ensure your training and test features are float64 for consistent model training
X_train_float = X_train.astype('float64')
X_test_float = X_test.astype('float64')

# ----------------- Model Initialization -----------------
reg_models = {
    "XGBoost Regressor": xgb.XGBRegressor(
        objective='reg:squarederror',
        n_estimators=100,
        random_state=42
    ),
    "Random Forest Regressor": RandomForestRegressor(
        n_estimators=100,
        random_state=42
    ),
    "Linear Regression": LinearRegression()
}

# ----------------- Training and Evaluation Loop -----------------
print("\n" + "="*60)
print("  STARTING REGRESSION MODEL TRAINING AND EVALUATION")
print("="*60)

results = {}

for name, model in reg_models.items():
    print(f"\n--- Training {name} ---")

    # Train using the standard (non-resampled) training data
    model.fit(X_train_float, y_reg_train)
    
    # Predict on the original test set
    y_reg_pred = model.predict(X_test_float)
    
    # Evaluate performance
    mae = mean_absolute_error(y_reg_test, y_reg_pred)
    rmse = np.sqrt(mean_squared_error(y_reg_test, y_reg_pred))
    r2 = r2_score(y_reg_test, y_reg_pred)
    
    results[name] = {'MAE': mae, 'RMSE': rmse, 'R2': r2}

    print(f"\n{name} Performance on Test Set:")
    print(f"  R-squared (R2): {r2:.4f} (Closer to 1.0 is better)")
    print(f"  Mean Absolute Error (MAE): {mae:.2f}")
    print(f"  Root Mean Squared Error (RMSE): {rmse:.2f}")
    print("-" * 50)
```

```python
import pandas as pd
import numpy as np

# --- ASSUMPTION ---
# 1. Models are trained and stored here:
# best_classifier = ... (e.g., models["XGBoost"])
# best_regressor = ... (e.g., reg_models["XGBoost Regressor"])

# 2. Label Encoder mapping for Classification Target:
# Assuming the classification model uses Label Encoding (0=Eligible, 1=High_Risk, 2=Not_Eligible)
# You MUST verify this mapping from your earlier LabelEncoder output (le.classes_).
CLASS_MAPPING = {0: 'Eligible', 1: 'High_Risk', 2: 'Not_Eligible'}

def predict_loan_decision(applicant_data: pd.DataFrame, requested_emi: float) -> dict:
    """
    Applies the dual-model strategy to predict eligibility and max loan amount.

    Args:
        applicant_data: A DataFrame containing the single applicant's features (X).
                        MUST be in the same format and column order as the training data.
        requested_emi: The monthly EMI amount requested by the applicant.

    Returns:
        A dictionary containing the final decision and predicted values.
    """
    
    # 1. Standardize Data Type
    # Convert features to float64, matching the type used for training/resampling
    applicant_data_float = applicant_data.astype('float64')

    # --- 2. CLASSIFICATION GATE CHECK ---
    
    # Predict eligibility (returns 0, 1, or 2)
    eligibility_encoded = best_classifier.predict(applicant_data_float)[0]
    eligibility_status = CLASS_MAPPING[eligibility_encoded]

    if eligibility_status == 'Eligible':
        
        # --- 3. REGRESSION AFFORDABILITY CHECK ---
        
        # Predict the maximum affordable monthly EMI amount
        max_emi_prediction = best_regressor.predict(applicant_data_float)[0]
        
        # Ensure prediction is non-negative and round to the nearest whole number
        max_emi_prediction = max(0, round(max_emi_prediction))

        # --- 4. FINAL DECISION LOGIC ---
        if max_emi_prediction >= requested_emi:
            final_decision = "Approved"
            recommendation = f"Loan approved. Recommended EMI is {requested_emi:.2f}. Max affordable EMI: {max_emi_prediction:.2f}"
        else:
            final_decision = "Conditional Approval"
            recommendation = f"Requested EMI ({requested_emi:.2f}) is too high. Max affordable EMI is {max_emi_prediction:.2f}. Offer a revised loan."
        
        return {
            "status": final_decision,
            "eligibility_check": eligibility_status,
            "predicted_max_emi": max_emi_prediction,
            "recommendation": recommendation
        }
        
    else:
        # --- 5. REJECTION ---
        return {
            "status": "Rejected",
            "eligibility_check": eligibility_status,
            "predicted_max_emi": None,
            "recommendation": f"Loan rejected. Applicant categorized as {eligibility_status}."
        }
```
